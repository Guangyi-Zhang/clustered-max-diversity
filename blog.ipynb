```python
%load_ext autoreload
%autoreload 2

import os
from os.path import join
from data import gen_topics_from_docs, gen_topics
```

```python
datadir = join('dataset','blogs')

files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(datadir):
    for file in f:
        if '.xml' in file:
            files.append(os.path.join(r, file))
```

```python
len(files)
```

```python
docs = []
for f in files:
    with open(f, 'r') as fin:
        try:
            data = fin.read()
        except:
            print(f)
            continue
        beg = data.find('<post>') + 6
        end = data.find('</post>')
        doc = data[beg:end].strip()
        docs.append(doc)
```

```python
len(docs)
```

```python
gen_topics_from_docs('blogs', docs, ntopic=10, passes=20, topic_threshold=0.25)
```
